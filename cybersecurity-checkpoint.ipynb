```python
import pandas as pd

df= pd.read_csv(r"C:\Users\veera\Downloads\GUIDE_Train.csv.zip", nrows=10000)
df.to_csv("guide_sample.csv", index=False)

print(df.shape)
print(df.columns)
print(df.dtypes)

```

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score, confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df
```

```python
# drop columns
df= df.drop(columns=['Id', 'IncidentId', 'AlertId', 'OrgId', 'DetectorId'])
```

```python
# Parse and Extract Date Features
df['Timestamp'] = pd.to_datetime(df['Timestamp'], errors='coerce')
df['Hour'] = df['Timestamp'].dt.hour
df['DayOfWeek'] = df['Timestamp'].dt.dayofweek
df['IsWeekend'] = df['DayOfWeek'].apply(lambda x: 1 if x >= 5 else 0)
df = df.drop(columns=['Timestamp'])  # drop original timestamp
```

```python
# convert to numeric
from sklearn.preprocessing import LabelEncoder

le_grade = LabelEncoder()
df['IncidentGrade'] = le_grade.fit_transform(df['IncidentGrade'])
# Mapping: 0 = BenignPositive, 1 = FalsePositive, 2 = TruePositive
```

```python
#hadling missing value
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = df[col].fillna('missing')
    else:
        df[col] = df[col].fillna(df[col].median())
```

```python
#Encoding categorical features
cat_cols = df.select_dtypes(include='object').columns.tolist()
print(cat_cols)
```

```python
from sklearn.preprocessing import LabelEncoder

for col in cat_cols:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col].astype(str))
```

```python
X = df.drop(columns=['IncidentGrade'])
y = df['IncidentGrade']
```

```python
from sklearn.model_selection import train_test_split

X_train, X_val, y_train, y_val = train_test_split(
    X, y, test_size=0.2, stratify=y, random_state=42
)
```

```python
# 1. Imports
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import classification_report, confusion_matrix
from sklearn.metrics import precision_score, recall_score, f1_score

# 2. Initialize the Random Forest model
rf_model = RandomForestClassifier(
    n_estimators=100,
    max_depth=None,
    class_weight='balanced',  # to handle imbalance
    random_state=42,
    n_jobs=-1
)

# 3. Fit the model
rf_model.fit(X_train, y_train)

# 4. Make predictions
y_pred = rf_model.predict(X_val)

# 5. Evaluate
print("üîç Classification Report:\n")
print(classification_report(y_val, y_pred, target_names=le_grade.classes_))

# Optional: calculate macro metrics manually
precision = precision_score(y_val, y_pred, average='macro')
recall = recall_score(y_val, y_pred, average='macro')
f1 = f1_score(y_val, y_pred, average='macro')

print(f"üéØ Macro Precision: {precision:.4f}")
print(f"üéØ Macro Recall: {recall:.4f}")
print(f"üéØ Macro F1 Score: {f1:.4f}")

# 6. Confusion Matrix
import seaborn as sns
import matplotlib.pyplot as plt

cm = confusion_matrix(y_val, y_pred)
plt.figure(figsize=(6, 5))
sns.heatmap(cm, annot=True, fmt='d', xticklabels=le_grade.classes_, yticklabels=le_grade.classes_, cmap="Blues")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.title("Confusion Matrix")
plt.show()
```
